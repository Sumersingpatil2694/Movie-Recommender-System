```python
import ast

def convert3(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:  
            L.append(i['name'])  
            counter += 1
        else:
            break
    return L
```

```python
import ast

def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L  
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x: x[1])[1:6]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distence = similarity[movie_index]
    movies_list = sorted(list(enumerate(distence)), reverse=True, key=lambda x: x[1])[1:6]
    
    print("Top 5 recommended movies similar to", movie, ":")
    for i in movies_list:
        recommended_movie_index = i[0]
        print(new_df.iloc[recommended_movie_index]['title'])  
```
